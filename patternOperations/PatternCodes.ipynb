### Frquent Patterns

```python
def frequent_words(Text, k):
    freq_map = {}
    n = len(Text)
    for i in range(n - k + 1):
        pattern = Text[i:i + k]
        if pattern in freq_map:
            freq_map[pattern] += 1
        else:
            freq_map[pattern] = 1
    max_count = max(freq_map.values())
    return [pattern for pattern, count in freq_map.items() if count == max_count]
```

### Find Patterns Forming Clumps in a String
Problem 12 @ FA2025 UTA Bioinformatics 
Given integers L and t, a string Pattern forms an (L, t)-clump inside a (larger) string Genome if there is an interval of Genome of length L in which Pattern appears at least t times. For example, TGCA
 forms a (25,3)-clump in the following Genome: gatcagcataagggtcccTGCAATGCATGACAAGCCTGCAgttgttttac
.

Clump Finding Problem
Find patterns forming clumps in a string.

Given: A string Genome, and integers k, L, and t.

Return: All distinct k-mers forming (L, t)-clumps in Genome.

Sample Dataset
CGGACTCGACAGATGTGAAGAAATGTGAAGACTGAGTGAAGAGAAGAGGAAACACGACACGACATTGCGACATAATGTACGAATGTAATGTGCCTATGGC
5 75 4
Sample Output
CGACA GAAGA AATGT

```python
def clump(Genome, k, L, t):
    clumps = set()
    n = len(Genome)
    for i in range(n - L + 1):
        window = Genome[i:i + L]
        freq_map = {}
        for j in range(L - k + 1):
            pattern = window[j:j + k]
            if pattern in freq_map:
                freq_map[pattern] += 1
            else:
                freq_map[pattern] = 1
        for pattern, count in freq_map.items():
            if count >= t:
                clumps.add(pattern)
    return clumps

if __name__ == "__main__":

    with open("rosalind_ba1e.txt", "r") as f:
        lines = f.read().strip().split("\n")
        Genome = lines[0]
        k, L, t = map(int, lines[1].split())
        #print(frequent_words(Genome, k))
        for c in (clump(Genome, k, L, t)):
            print(c, end=" ")
    
```

```python
def minimum_skew(Genome):
    skew = 0
    min_skew = 0
    min_positions = []
    for i, nucleotide in enumerate(Genome):
        if nucleotide == 'C':
            skew -= 1
        elif nucleotide == 'G':
            skew += 1
        if skew < min_skew:
            min_skew = skew
            min_positions = [i + 1]
        elif skew == min_skew:
            min_positions.append(i + 1)
    return min_positions

if __name__ == "__main__":

    # genome = "CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG"
    # print(minimum_skew(genome))
    with open("rosalind_ba1f.txt", "r") as f:
        genome = f.read().strip()
        for pos in minimum_skew(genome):
            print(pos, end=" ")
```

  <h1 style="margin-top:1rem;">Hamming Distance Problem</h1>

  <h2>Definition</h2>
  <p>
    The <strong>Hamming distance</strong> between two strings of equal length is the number of positions
    at which the corresponding symbols are different. In bioinformatics, this commonly measures
    dissimilarity between DNA, RNA, or protein sequences.
  </p>

  <h2>Formal Description</h2>
  <p>
    Given two strings <em>p = p<sub>1</sub>p<sub>2</sub>…p<sub>k</sub></em> and
    <em>q = q<sub>1</sub>q<sub>2</sub>…q<sub>k</sub></em> of length <em>k</em>:
  </p>
  <ul>
    <li>A <strong>mismatch</strong> occurs at position <em>i</em> if <code>p<sub>i</sub> ≠ q<sub>i</sub></code>.</li>
    <li>The <strong>Hamming distance</strong> is the total number of mismatches.</li>
  </ul>
  <div style="background:#f7f7f9; border:1px solid #e1e1e8; padding:0.8rem; border-radius:8px; overflow:auto;">
    <code>
      HammingDistance(p, q) = Σ<sub>i=1..k</sub> [p<sub>i</sub> ≠ q<sub>i</sub>]
    </code>
    <p style="margin:0.4rem 0 0 0; font-size:0.95em; color:#555;">
      where [·] is 1 if the statement is true, otherwise 0.
    </p>
  </div>

  <h2>Example</h2>
  <pre style="background:#0b1021; color:#e6e6e6; padding:0.9rem; border-radius:8px; overflow:auto;"><code>p = GGGCCGTTGGT

  <h2>Applications</h2>
  <ul>

    <li>Pattern matching with mismatches in genome analysis.</li>

  <h1 style="margin-top:1rem;">d-Neighborhood of a String</h1>

  <h2>What is a d-Neighborhood?</h2>
  <p>
    For a DNA <em>k</em>-mer <strong>Pattern</strong> and a non-negative integer <strong>d</strong>, the
    <strong>d-neighborhood</strong>, written as <code>Neighbors(Pattern, d)</code>, is the set of all
    <em>k</em>-mers whose <a href="#hamming" style="text-decoration:none;">Hamming distance</a> from
    <strong>Pattern</strong> is at most <strong>d</strong> (i.e., they differ in ≤ d positions).
  </p>

  <div id="hamming" style="background:#f7f7f9; border:1px solid #e1e1e8; padding:0.8rem; border-radius:8px;">
    <strong>Hamming distance</strong> between equal-length strings is the number of positions where they differ.
  </div>

  <h2>Intuition</h2>
  <ul>
    <li>If <strong>d = 0</strong> → the neighborhood contains only the pattern itself.</li>
    <li>If <strong>d = 1</strong> → include all strings formed by changing <em>any one</em> position to any other nucleotide.</li>
    <li>If <strong>d = 2</strong> → include all strings that differ in 0, 1, or 2 positions, and so on.</li>
  </ul>

  <h2>Example</h2>
  <p><strong>Pattern</strong> = <code>ACG</code>, <strong>d</strong> = <code>1</code></p>

  <h3 style="margin-bottom:0.4rem;">Change exactly one position:</h3>
  <ol>
    <li>
      Change the first character (<code>A</code> → <code>C,G,T</code>):
      <code>CCG</code>, <code>GCG</code>, <code>TCG</code>
    </li>
    <li>
      Change the second character (<code>C</code> → <code>A,G,T</code>):
      <code>AAG</code>, <code>AGG</code>, <code>ATG</code>
    </li>
    <li>
      Change the third character (<code>G</code> → <code>A,C,T</code>):
      <code>ACA</code>, <code>ACC</code>, <code>ACT</code>
    </li>
  </ol>

  <div style="background:#0b1021; color:#e6e6e6; padding:0.9rem; border-radius:8px; overflow:auto;">
    <pre style="margin:0;"><code>Neighbors(ACG, 1) =
CCG
TCG
GCG
AAG
ATG
AGG
ACA
ACC
ACT</code></pre>
  </div>

  <h2>Why it matters</h2>
  <ul>
    <li><strong>Motif finding:</strong> capture conserved patterns that tolerate a few mutations.</li>
    <li><strong>Approximate matching:</strong> search for patterns allowing mismatches in noisy biological data.</li>
    <li><strong>Robust analysis:</strong> accounts for sequencing errors and natural variation.</li>
  </ul>

  <h2>How to think algorithmically (no code)</h2>
  <ul>
    <li><strong>Enumerate by positions:</strong> choose 0..d positions to mutate; for each chosen position, substitute with any of the other nucleotides.</li>
    <li><strong>Recursive idea:</strong> build neighbors of the suffix, then prepend either the original first symbol (keeping the mismatch count) or alternative symbols (consuming one mismatch).</li>
    <li><strong>De-duplicate:</strong> collect results in a set to avoid repeats.</li>
  </ul>
</section>


```python
def d_Neighbors(Pattern, d):
    if d == 0:
        return {Pattern}
    if len(Pattern) == 1:
        return {'A', 'C', 'G', 'T'}
    
    Neighborhood = set()
    SuffixNeighbors = d_Neighbors(Pattern[1:], d)
    for text in SuffixNeighbors:
        if hamming_distance(Pattern[1:], text) < d:
            for nucleotide in 'ACGT':
                Neighborhood.add(nucleotide + text)
        else:
            Neighborhood.add(Pattern[0] + text)
    return Neighborhood

if __name__ == "__main__":


    # pattern = "ACG"
    # d = 1
    # print(' '.join(d_Neighbors(pattern, d)))
    with open('rosalind_ba1n.txt', 'r') as f:
        lines = f.read().strip().split('\n')
        Pattern = lines[0]
        d = int(lines[1])
        neighbors = d_Neighbors(Pattern, d)
        print(' '.join(neighbors))
```

<p>

</p>
<ul>
  <li>A <strong>mismatch</strong> means a difference in characters at the same position (i.e., Hamming distance).</li>
  <li>We slide a window of the same length as the pattern across the text.</li>
  <li>At each position, we compare the substring with the pattern using the <strong>Hamming Distance</strong>.</li>
</ul>

```python

def approximate_pattern_matching(Pattern, Text, d):
    """
    Find all starting positions where Pattern appears in Text
    with at most d mismatches (Hamming distance ≤ d).
    """
    positions = []
    k = len(Pattern)
    for i in range(len(Text) - k + 1):
        substring = Text[i:i + k]
        if hamming_distance(Pattern, substring) <= d:
            positions.append(i)
    return positions

if __name__ == "__main__":
    
    with open('rosalind_ba1h.txt', 'r') as f:
        lines = f.read().strip().split('\n')
        Pattern = lines[0]
        Text = lines[1]
        d = int(lines[2])
        positions = approximate_pattern_matching(Pattern, Text, d)
        print(' '.join(map(str, positions)))
```

  <h1 style="margin:0 0 10px">Frequent Words with Mismatches</h1>
  <p style="color:#374151; margin:0 0 16px">
    We generalize “Most Frequent k-mers” to allow up to <code>d</code> mismatches.
    For each candidate k-mer <code>Pattern</code>, we count how often it occurs in <code>Text</code>
    within Hamming distance ≤ <code>d</code>.
  </p>

  <h2 style="margin:18px 0 8px">Key Definitions</h2>
  <ul style="color:#374151; margin:0 0 16px">
    <li><strong>Hamming distance</strong> <code>Hamming(p, q)</code>: number of positions where strings <code>p</code> and <code>q</code> (same length) differ.</li>
    <li><strong>Count<sub>d</sub>(Text, Pattern)</strong>: number of starting positions <code>i</code> where <code>Text[i..i+k-1]</code> and <code>Pattern</code> differ in ≤ <code>d</code> positions.</li>
    <li><strong>Neighborhood</strong> (or <strong>d-neighborhood</strong>) of a k-mer: the set of all k-mers whose Hamming distance from the k-mer is ≤ <code>d</code>.</li>
    <li><strong>Goal</strong>: find all k-mers maximizing <code>Count<sub>d</sub>(Text, Pattern)</code>.</li>
  </ul>

  <h2 style="margin:18px 0 8px">Why Neighborhoods?</h2>
  <p style="color:#374151; margin:0 0 12px">
    If a substring <code>S</code> of <code>Text</code> occurs at position <code>i</code>, then any <code>Pattern</code> within ≤ <code>d</code> mismatches of <code>S</code> is a valid “hit” at <code>i</code>.
    So, instead of matching every possible <code>Pattern</code> to the whole <code>Text</code>, we:
  </p>
  <ol style="color:#374151; margin:0 0 16px">
    <li>Slide a window of size <code>k</code> over <code>Text</code> to get each k-mer <code>S</code>.</li>
    <li>Generate the <strong>d-neighborhood</strong> of <code>S</code> (all k-mers within ≤ <code>d</code> mismatches of <code>S</code>).</li>
    <li>Increase the count for each neighbor. The most frequent neighbors are the answer.</li>
  </ol>

  <h2 style="margin:18px 0 8px">Example</h2>
  <div style="padding:12px 14px; background:#f9fafb; border:1px solid #e5e7eb; border-radius:10px; color:#374151; margin-bottom:14px">
    <p style="margin:0 0 6px"><strong>Sample Dataset</strong></p>
    <pre style="margin:0; padding:10px; background:#0b1020; color:#e5e7eb; border-radius:8px; overflow:auto">Text = ACGTTGCATGTCGCATGATGCATGAGAGCT
k = 4, d = 1</pre>
  </div>
  <p style="color:#374151; margin:0 0 10px">
    The most frequent 4-mers with ≤1 mismatch are:
  </p>
  <pre style="margin:0 0 16px; padding:10px 12px; background:#0b1020; color:#e5e7eb; border-radius:8px; overflow:auto">GATG ATGC ATGT</pre>

  <h2 style="margin:18px 0 8px">Algorithm (Neighborhood Counting)</h2>
  <ol style="color:#374151; margin:0 0 16px">
    <li>Initialize a frequency map <code>freq</code> (dictionary) to 0.</li>
    <li>For each index <code>i</code> from <code>0</code> to <code>|Text|-k</code>:
      <ul>
        <li>Let <code>S = Text[i..i+k-1]</code>.</li>
        <li>Compute <code>Neighbors(S, d)</code> (the d-neighborhood of <code>S</code>).</li>
        <li>For each <code>N</code> in <code>Neighbors(S, d)</code>, do <code>freq[N] += 1</code>.</li>
      </ul>
    </li>
    <li>Let <code>m = max(freq.values())</code>.</li>
    <li>Return all keys <code>Pattern</code> where <code>freq[Pattern] == m</code>.</li>
  </ol>

  <h2 style="margin:18px 0 8px">Neighbors Recurrence (Concept)</h2>
  <p style="color:#374151; margin:0 0 12px">
    To build <code>Neighbors(pattern, d)</code> recursively:
  </p>
  <ul style="color:#374151; margin:0 0 16px">
    <li>If <code>d == 0</code>, return <code>{pattern}</code>.</li>
    <li>If <code>|pattern| == 1</code>, return <code>{A, C, G, T}</code>.</li>
    <li>Recurse on the suffix: <code>Neighbors(pattern[1:], d)</code>.
      For each suffix neighbor:
      <ul>
        <li>If <code>Hamming(pattern[1:], suffixNeighbor) &lt; d</code>, you may place any nucleotide in front.</li>
        <li>Else, you must keep the original first nucleotide.</li>
      </ul>
    </li>
  </ul>

  <h2 style="margin:18px 0 8px">Complexity</h2>
  <p style="color:#374151; margin:0 0 12px">
    Let <code>n = |Text|</code>. We have <code>n - k + 1</code> windows. The size of a d-neighborhood is
    roughly <code>∑<sub>i=0..d</sub> C(k, i) · 3<sup>i</sup></code>.
    This is practical for small <code>k</code> and small <code>d</code> (e.g., <code>k ≤ 10</code>, <code>d ≤ 2–3</code>).
  </p>

  <h2 style="margin:18px 0 8px">Edge Cases & Tips</h2>
  <ul style="color:#374151; margin:0 0 16px">
    <li>If <code>d = 0</code>, the problem reduces to exact k-mer counting.</li>
    <li>If <code>k &gt; |Text|</code>, answer is empty (no k-mers fit).</li>
    <li>DNA alphabet is assumed to be <code>{A, C, G, T}</code>.</li>
    <li>To speed up, memoize <code>Neighbors()</code> for repeated substrings.</li>
  </ul>

```python
from collections import defaultdict

def hamming_distance(p, q):
    """Return the Hamming distance between equal-length strings p and q."""
    return sum(pi != qi for pi, qi in zip(p, q))

def neighbors(pattern, d):
    """
    Generate the d-neighborhood of a pattern:
    all strings within d mismatches of 'pattern'.
    """
    if d == 0:
        return {pattern}
    if len(pattern) == 1:
        return {"A", "C", "G", "T"}
    
    neighborhood = set()
    suffix_neighbors = neighbors(pattern[1:], d)
    for text in suffix_neighbors:
        if hamming_distance(pattern[1:], text) < d:
            for nucleotide in "ACGT":
                neighborhood.add(nucleotide + text)
        else:
            neighborhood.add(pattern[0] + text)
    return neighborhood

def frequent_words_with_mismatches(text, k, d):
    """
    Find the most frequent k-mers with at most d mismatches in 'text'.
    """
    freq_map = defaultdict(int)
    
    # loop through all k-mers in the text
    for i in range(len(text) - k + 1):
        pattern = text[i:i+k]
        # generate all neighbors of this k-mer
        for neighbor in neighbors(pattern, d):
            freq_map[neighbor] += 1
    
    # find maximum count
    max_count = max(freq_map.values())
    
    # collect all k-mers with maximum frequency
    return [kmer for kmer, count in freq_map.items() if count == max_count]


# --------------------------
# Example from problem
# --------------------------
# text = "ACGTTGCATGTCGCATGATGCATGAGAGCT"
# k, d = 4, 1
# result = frequent_words_with_mismatches(text, k, d)
#print(" ".join(result))

with open("rosalind_ba1i.txt", "r") as f:
    lines = f.read().strip().split("\n")
    text = lines[0]
    k, d = map(int, lines[1].split())
    result = frequent_words_with_mismatches(text, k, d)
    print(" ".join(result))
```

```python
def reverse_complement(pattern):
    """Return the reverse complement of a DNA string."""
    complement = {"A":"T", "T":"A", "C":"G", "G":"C"}
    return "".join(complement[base] for base in reversed(pattern))

def frequent_words_with_mismatches_and_rc(text, k, d):
    freq_map = defaultdict(int)

    # generate all possible neighbors and their reverse complements
    for i in range(len(text) - k + 1):
        substring = text[i:i+k]
        for neighbor in neighbors(substring, d):
            rc = reverse_complement(neighbor)
            freq_map[neighbor] += 1
            freq_map[rc] += 0  # ensure RC is in the dict

    # adjust counts to include reverse complement frequencies
    combined_freq = defaultdict(int)
    for pattern in freq_map:
        combined_freq[pattern] = freq_map[pattern] + freq_map[reverse_complement(pattern)]

    # find maximum frequency
    max_count = max(combined_freq.values())
    return [pattern for pattern, count in combined_freq.items() if count == max_count]


if __name__ == "__main__":
    with open("rosalind_ba1j.txt", "r") as f:
        lines = f.read().strip().split("\n")
        text = lines[0]
        k, d = map(int, lines[1].split())
        result = frequent_words_with_mismatches_and_rc(text, k, d)
        print(" ".join(result))
```
